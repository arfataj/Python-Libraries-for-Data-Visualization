```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
np.pi
```

```python
#compute x and y coordinate for points in sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
#subplot
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up subplot grid that has height 2 and width 1
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cosine')

plt.show()
```
