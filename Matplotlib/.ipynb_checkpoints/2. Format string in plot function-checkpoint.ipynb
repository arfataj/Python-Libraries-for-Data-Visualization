# Matplotlib Tutorial: pyplot.plot format strings and other options

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+')
```

```python
# Same effect as 'g+' format string
plt.plot(x,y,color='red',marker='+',linestyle='dotted')
```

# markersize

```python
plt.plot(x,y,color='red',marker='D',linestyle='',markersize=10)
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'g+-.')
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,'rD--')
```

# "alpha" property to control transparency of line chart

```python
plt.plot(x,y,color='green',alpha=1)  # alpha can be specified on a scale 0 to 1
```

```python
plt.plot(x,y,color='green',alpha=0.5)
```

```python
plt.plot(x,y,color='green',alpha=0.2)
```

```python
plt.plot(x,y,'g<',alpha=0.5) 
```

```python
plt.plot(x,y,color='black',alpha=0) #doesn't display
```
