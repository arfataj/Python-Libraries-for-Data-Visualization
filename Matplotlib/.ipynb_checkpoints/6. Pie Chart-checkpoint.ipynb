```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Draw pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.pie(exp_vals,labels=exp_labels)
```

# Draw pie chart as perfect circle 

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

# Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', radius=1.5)
plt.show()
```

# Explode

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', radius=1.5, explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', radius=1.5, explode=[0,0.2,0.1,0,0])
plt.show()
```

# counterclock and angle properties

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', radius=1.5, explode=[0,0.2,0.1,0,0], startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%', radius=1.5, explode=[0,0.2,0.1,0,0], startangle=180)
plt.show()
```
